### 时间融合：把事件发生时间和蜂鸣器时间按顺序排列

```python
import pandas as pd
import time
# 蜂鸣器时间
fmq_data=pd.read_excel('D:/data/fmq_data.xlsx')
fmq_data = fmq_data.T
fmq_data.columns = ['time']
fmq_data['type'] = 21
fmq_data
```

```python
# 事件时间
sj_data=pd.read_excel('D:/data/sj_data.xlsx')
sj_data
```

```python
sj_data['datatime'] = sj_data['y_1'].map(str)  +'-'+ sj_data['y_2'].map(str) + '-' + sj_data['y_3'].map(str)
sj_data['sfm'] = sj_data['y_4'].map(str)+':'+sj_data['y_5'].map(str)+':'+round(sj_data['y_6'],3).map(str)
sj_data['time'] = sj_data['datatime'] + ' '+ sj_data['sfm']
sj_data = sj_data.rename(columns={'y_7':'type'})
sj_data = sj_data.loc[:,['time','type']]
sj_data['time'] = pd.to_datetime(sj_data['time'])
sj_data
```

```python
data = pd.concat([fmq_data,sj_data],ignore_index=True)
data = data.sort_values(by='time')
data
```

```python
data.to_csv('D:/data/时间融合.csv',index = False,encoding="utf-8")
```

### 时间对齐：把事件时间、键盘时间、EEG上时间对齐

```python
# 读取志愿者敲键盘的时间
jp_data = pd.read_csv('../data/0803_fujunwen/1_1/1_1_0803_fujunwen.txt', sep='\t', header=None)
jp_data.columns = ['time_jp']
jp_data['time_jp'] = pd.to_datetime(jp_data['time_jp'])
# 敲键盘的第一次和第二次的时间差，也就是我们敲键盘的第一下和志愿者敲键盘的第一下的时间差
c = (jp_data['time_jp'][1] - jp_data['time_jp'][0]).seconds
jp_data
```

```python
c
```

```python
## 读取事件发生的时间
sj_data = pd.read_csv('../data/时间融合.csv')
sj_data = sj_data.rename(columns = {'time':'time_sj'})
sj_data
```

```python
## 两者对应
## 先把事件分成两组：危险事件+蜂鸣器、不危险事件
sj_data_wx = sj_data[sj_data['type'].isin([1,2,5,6,9,10,13,14,19,20,21])]
sj_data_bwx = sj_data[sj_data['type'].isin([3,4,7,8,11,12,15,16,17,18])]
sj_data_wx = sj_data_wx.reset_index(drop=True)
sj_data_bwx = sj_data_bwx.reset_index(drop=True)
sj_data_wx
```

```python
sj_data_bwx
```

### 危险事件+蜂鸣器

```python
## 匹配危险事件时间后面最近的键盘时间
sj_data_wx['time_sj'] = pd.to_datetime(sj_data_wx['time_sj'])

jp_data['time_jp'] = pd.to_datetime(jp_data['time_jp'])
time_merge = pd.merge_asof(sj_data_wx, jp_data,left_on = 'time_sj',right_on = 'time_jp', direction='forward')
time_merge
```

```python
# 键盘上的第一次对应eeg图像上的第一个marker 12.04s，
time_jd = time_merge.copy()
time_jd.loc[0,'time_eeg'] = c+12.04
## 事件发生时间和键盘反应时间之间的差，并转化为秒
time_jd.loc[:,'difference'] = (time_jd['time_jp'] - time_jd['time_sj']).dt.total_seconds()
# 推算后面的数据
for i in range(1,len(time_jd)):
    time_jd.loc[i,'time_eeg'] = c+ 12.04 + (time_jd.loc[i,'time_jp'] -time_jd.loc[0,'time_jp']).seconds
time_jd['time_eeg'] = time_jd['time_eeg'].astype('float64')
## 在eeg上对应的真实事件时间
time_jd['time_eeg_real'] = time_jd['time_eeg'] - time_jd['difference']
time_jd
```

```python
time_jd.to_csv('D:/data/危险事件.csv',index = False,encoding="utf-8")
```

```python
## 只保留这两列
data = time_jd.loc[:,['time_eeg_real','type']]
data = data.rename(columns={'time_eeg_real':'latency'})
data
```

### 不危险事件

```python
sj_data_bwx['time_sj']  = pd.to_datetime(sj_data_bwx['time_sj'])
sj_data_bwx.loc[:,'time_eeg_real'] = time_jd['time_eeg_real'][0] + (sj_data_bwx['time_sj']-time_jd['time_sj'][0]).dt.total_seconds()
sj_data_bwx
```

```python
data1 = sj_data_bwx.loc[:,['time_eeg_real','type']]
data1 = data1.rename(columns={'time_eeg_real':'latency'})
data1
```

### 汇总

```python
data2 = pd.concat([data,data1],axis=0)
data2 = data2.sort_values(by='latency')
data2 = data2.reset_index(drop=True)
data2['latency'] = data2['latency']
# 第一列保留三位小数
data2['latency'] = data2['latency'].apply(lambda x:round(x,3))
data2
```

```python
data2.to_csv('D:/output/event.txt',sep='\t',index=False)
```
